```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import interp1d
```

```python
dfM = pd.read_csv("milliqan300.csv",names=['x','y'])
dfO = pd.read_csv("moedal30.csv",names=['x','y'])
dfM['x2'] = dfO['x']
dfM['y2'] = dfO['y']
```

```python
ax1 = dfM.plot(kind='scatter', x='x', y='y', color='r',label='milliQan 300/fb')    
ax2 = dfM.plot(kind='scatter', x='x2', y='y2', color='g',label='moedal 30/fb', ax=ax1)
plt.ylabel("Q/e")
plt.xlabel("GeV")
plt.yscale("log")
plt.xscale("log")
```

```python
xvals = np.arange(0, 100, 0.01)

milli_chargevals = np.interp(xvals, dfM['x'], dfM['y'])
moedal_chargevals = np.interp(xvals, dfM['x2'], dfM['y2'])

ax1 = plt.scatter(xvals, milli_chargevals, s=1, label='data')
ax2 = plt.scatter(dfM['x'], dfM['y'], s=1, label='numpy')
plt.ylabel('Ratio of charges (milli/moedal) [Q/e]')
plt.xlabel('Mass [GeV]')
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.1, 100)

# xvals = np.arange(dfM['x'][0], dfM['x2'][71], 0.01)

# f = interp1d(dfM['x'], dfM['y'])
# f2 = interp1d(dfM['x2'], dfM['y2'])
# plt.scatter(xvals, f(xvals)/f2(xvals), s=1)
# plt.xscale('log')
# plt.yscale('log')
# plt.ylabel('Ratio of charges (milli/moedal) [Q/e]')
# plt.xlabel('Mass [GeV]')
# plt.xscale('log')
# plt.xlim(0.1, 100)
```
